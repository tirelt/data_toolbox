```python
df = df.drop_duplicates(keep='first') #drop duplicate
```

```python
df["Sex_nbr"]=df["Sex"].apply(lambda x: 1 if x== 'male' else 0) #create column
```

```python
df = df[(df["Age"] >= 12 ) & ~(pd.isnull(df["Age"]))] #remove babies an missing age
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=df["Fare"],histnorm="percent"))#,nbinsx=10
fig.update_layout(title='Histogram title')
fig.show()
```

```python
df[df['Fare'] >= 200].sort_values(by="Fare",ascending=False).dropna() #check outlier
```

```python
# preprocessing
df_preprocessed = df.copy()
```

```python
# for numerical columns, we replace the NaNs by the mean of the column
for feature_name in ['Age']:
    feature_mean = int(df[feature_name].mean())
    feature_missing = pd.isnull(df[feature_name]).sum()
    feature_missing_pct = feature_missing/len(df)
    print(f"{feature_missing_pct:.2%} missing for {feature_name}, {feature_missing} replaced by mean value = {feature_mean:.2f}")
    df_preprocessed[feature_name] = df_preprocessed[feature_name].fillna(feature_mean)
```

```python
# Read dataset
df = pd.read_csv("../data/tick_data/googl_trade_2023_05_12.txt",header=None) # changed index_col
df = df.set_axis(["timestamp","price","volume","exchange","cond"], axis=1)
df['timestamp'] = pd.to_datetime(df['timestamp']) 
df.head()
```

```python
df["hour"] = df["timestamp"].progress_apply(lambda x: x.hour)
```

```python
df["minute"]=  df["timestamp"].dt.minute
```

```python
df_candle = df.sort_values(by='timestamp').groupby(['minute'])['price'].apply(lambda x: {
    'open': x.iloc[0],
    'high': x.max(),
    'low': x.min(),
    'close': x.iloc[-1]
}).unstack(level=-1)
```

```python
fig = go.Figure()
fig.add_trace(go.Candlestick(
    x=df_candle.index,
    open=df_candle['open'],
    high=df_candle['high'],
    low=df_candle['low'],
    close=df_candle['close'],
))
fig.update_layout(title=f'Price')
fig.show()
```

```python
df_trade_by_min = df.groupby(['hour','minute'])['timestamp'].count()
df_trade_by_min = df_trade_by_min.reset_index(level=[0,1])
df_trade_by_min["minute_from_start"] = df_trade_by_min["hour"]*60+df_trade_by_min["minute"]
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_trade_by_min["minute_from_start"], y=df_trade_by_min['timestamp']))
fig.update_layout(title=f'NUmber of trades by hour')
fig.show()
```

```python
#hour seasonality
df_min_seas =df.groupby(['hour','minute'])['timestamp'].count().unstack(0)
df_min_seas["mean"]=df_min_seas.mean(axis=1)
df_min_seas["std"]=df_min_seas.std(axis=1)
```

```python
df_min_seas.index
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_min_seas.index, y=df_min_seas["mean"], error_y=dict(type='data', array=df_min_seas["std"], visible=True)))
fig.update_layout(title=f'NUmber of trades by hour')
fig.show()
```
